What day of the week is the busiest?

```python
daily_quantity = sales.groupby('day_of_week')['quantity_sold'].sum()
sns.barplot(x=daily_quantity.index, y=daily_quantity.values,order=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'])
plt.xlabel('Day of the Week')
plt.ylabel('Total Quantity Sold')
plt.title('Total Quantity Sold by Day of the Week')
plt.show()
```

From this barchart we see monday has the most amount of orders than the other days of the week. Monday Tuesday and Wednesday are all relatively close together and are more busy than the other days.

```python
monday_sales = sales[sales['day_of_week'] == 'Monday']
sns.displot(data=monday_sales, x='unit_price')
plt.xlabel('Drink Price')
plt.ylabel('Quantity Sold')
plt.title('Average Drink Price Sold on Monday')
plt.show()
```

```python
Friday_sales = sales[sales['day_of_week'] == 'Friday']
sns.displot(data=Friday_sales, x='unit_price')
plt.xlabel('Drink Price')
plt.ylabel('Quantity Sold')
plt.title('Average Drink Price Sold on Friday')
plt.show()
```

What is the relationship between calories and total fat in grams for the drinks and how does whip cream affect this as well?

```python
sns.scatterplot(data=nutritional,x='total_fat_g',y='calories',hue='whip')
plt.xlabel('Total Fat')
plt.ylabel('Calories')
plt.title('Relationship between Calories and Total Fat')
plt.show()
```

There is a positive relationship between how much total fat is in a drink and how many calories are in it as well.

When looking at whip or no whip you see a distinct distribution of groups. There are a few options where you can have no whip and get more total fat and calories whereas most of the time choosing no whip will lead to less calories and total fat.

What price of starbucks drink is purchased the most?

```python
sales_by_price = sales.groupby('unit_price')['total_sales'].sum()
sales_by_price.head(15)
sales_by_price.reset_index()
```

```python
sales_by_price_df = sales_by_price.reset_index()
print(sales_by_price_df.head())
```

```python
sns.scatterplot(data=sales_by_price_df, x='unit_price', y='total_sales')
plt.xlabel('Unit Price')
plt.ylabel('Total Sales')
plt.title('Relationship between Unit Price and Total Sales')
plt.show()
```

From this scatterplot I found that drinks priced between 4 dollars and 50 cents and 5 dollars had the most total sales.

What product has the most total sales?

```python
product_sales = sales.groupby('product_name')['total_sales'].sum()
product_sales.sort_values(ascending=False,inplace=True)
product_sales.head(15)
```

```python
top_15_products = product_sales.head(15)
plt.figure(figsize=(10, 8))
sns.barplot(x=top_15_products.values, y=top_15_products.index)
plt.xlabel('Total Sales')
plt.ylabel('Product Name')
plt.title('Top 15 Products by Total Sales')
plt.tight_layout()
plt.show()
```

When looking at the top 15 products of total sales, the Espresso Macchiato has the most total sales. It is very close to the total sales of the peach tranquility brewed tea.

When looking at the top 15 most popular drinks which have the most calories?

```python
top_15_products_df = top_15_products.reset_index()
top_15_products_df.columns = ['product_name', 'total_sales']
print(top_15_products_df.head())
```

```python
merged_df = pd.merge(top_15_products_df, nutritional, on='product_name', how='inner')
merged_df.head()
```

```python
product_calories = merged_df.groupby('product_name')['calories'].mean()
product_calories.sort_values(ascending=False,inplace=True)
product_calories.head(15)
```

```python
product_calories_df = product_calories.reset_index()
product_calories_df.columns = ['product_name', 'calories']
print(product_calories_df.head())
```

```python
sns.barplot(data=product_calories_df, x='calories', y='product_name')
plt.xlabel('Calories')
plt.ylabel('Product Name')
plt.tight_layout()
plt.title('Calories by Product Name')
plt.show()
```

When looking at the top 15 products calories we can see that the Caramel Apple Spice drink has the highest average calorie count out of theose drinks. In contrary all of the tea drinks have zero calories.
